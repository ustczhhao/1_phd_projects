```python
'''
Here we analyzed the best copy number of beneficial mutations in FGM. We generated 10000 beneficial mutations for the ancestor
which has a fitness of ~0.5, and then for each beneficial mutation, we investigated the fitness with 1 copy, 2 copies, ..., till
MAC ploidy copies of each beneficial mutations. We then found the copy number that can lead to the highest fitness (best copy 
number).

1. We got the best copy number for all the 10000 beneficial mutations, and then calculated the mean and STD of them.

2. We got the best copy number for all the 10000 beneficial mutations, and then investigated the distribution of these copy number
(i.e., for the 10000 beneficial mutations, we calculated how many beneficial mutations have such number as their
best copy number for each value ranging from 1 to MAC ploidy)

3. For all the 10000 beneficial mutations generated, we calculated and stored the mean fitness of them when they are present as
1 copy, 2 copies, ..., MAC ploidy copies, and then see how the mean fitness changes as the copy number changes. Also we searched
the copy number that leads to the highest mean fitness of the 10000 beneficial mutations.

'''
```

```python
def read_file(data):
    with open(data, 'rb') as f:
        x = pickle.load(f)
        
    return x
```

```python
copy_num_file = ['Total_Max_CopyNum_P1', 'Total_Max_CopyNum_P2', 'Total_Max_CopyNum_P4', 'Total_Max_CopyNum_P8', \
                 'Total_Max_CopyNum_P16', 'Total_Max_CopyNum_P32', 'Total_Max_CopyNum_P45', 'Total_Max_CopyNum_P64', \
                 'Total_Max_CopyNum_P128']
```

```python
total_max_cn = []

for i in copy_num_file:
    each_p_cn = read_file(i)
    
    total_max_cn.append(each_p_cn)
```

```python
len(total_max_cn[7])
```

```python
ploidy_list = [1,2,4,8,16,32,45,64,128]
```

```python
total_max_cn = np.array(total_max_cn)
```

```python
total_cn_distribution = []

for i in range(len(total_max_cn)):
    each_p_cn_distribution = []
    for j in range(1, ploidy_list[i]+1):
        each_p_cn_distribution.append(len(total_max_cn[i][total_max_cn[i]==j]))
        
    total_cn_distribution.append(each_p_cn_distribution)
```

```python
max_cn_mean = []
max_cn_std = []

for i in total_max_cn:
    print len(i)
    max_cn_mean.append(np.nanmean(i))
    max_cn_std.append(np.nanstd(i))
```

```python
def publication_figure(axis):
    plt.rcParams.update({'font.size': 30})
    ax = axis
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_linewidth(6)
    ax.spines['bottom'].set_linewidth(6)
    ax.spines['left'].set_position(('outward', 10))
    ax.spines['bottom'].set_position(('outward', 10))
    #ax.xaxis.set_ticks([])
    ax.tick_params(axis='x', which='both', bottom='on', top='off',labelbottom='on', size=20, width=6, direction='out', pad=50, colors='black')
    ax.tick_params(axis='y', which='both', left='on', right='off',labelbottom='on', size=20, width=6, direction='out', pad=50, colors='black')
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

fig = plt.figure(figsize=(16, 12), dpi = 300)
ax = fig.add_subplot(111)
publication_figure(ax)

ax.set_ylim(0, 70)
ax.set_xlim(1.5,150)

ax.set_xscale('log',basex=2)

x = [2, 4, 8, 16, 32, 45, 64, 128]
a = max_cn_mean[1:]
error_a = 1.96*(np.array(max_cn_std)[1:])/(10000**0.5)

b = np.array(x)/2



ax.errorbar(x, a, yerr = error_a, label = 'Mean Best Copy Number', fmt='o', color = 'b', markersize =14, lw=2, capsize=10, capthick=2)
ax.plot(x,a, color = 'b',ls = '--', alpha = 0.8)

ax.plot(x,b, label = 'Ploidy/2', marker = '^', color = 'dimgrey',ls = '--',markersize =14,  alpha = 0.8)


ax.set_xlabel('Ploidy', fontsize =36)
ax.set_ylabel('Mean Best Copy Number', fontsize =36)

ax.set_xticks([2,4,8, 16, 32, 45, 64, 128]) # choose which x locations to have ticks
ax.set_xticklabels([2,4,8, 16, 32, 45, 64, 128]) # set the labels to display at those ticks

plt.legend(loc='upper left', frameon=False, prop={'size':30})
plt.show()
```

```python
file_200 = ['T7_Fitness_Amito_200_1K_E1.csv', 'T7_Fitness_Mito_200_1K_E1.csv', 'T7_Fitness_SF_200_1K_E1.csv', \
         'T7_Fitness_RM_200_1K_E1.csv', 'T7_Fitness_SF_E100_200_1K_E1.csv', 'T7_Fitness_RM_E100_200_1K_E1.csv']

mfit_200 = []
log_mfit_200 = []

varfit_200 = []
varlog_fit_200 =[]

for i in file_200:
    f = pd.read_csv(i,dtype=str)
    
    mfit = []
    log_mfit = []
    
    varfit = []
    varlog_fit =[]
    
    mfit.extend(list(pd.Series(f.meanFit,dtype=float)))
    log_mfit.extend(list(pd.Series(f.log_meanFit,dtype=float)))
    
    varfit.extend(list(pd.Series(f.varFit,dtype=float)))
    varlog_fit.extend(list(pd.Series(f.varlog_Fit,dtype=float)))
    
    mfit_200.append(mfit)
    log_mfit_200.append(log_mfit)
    
    varfit_200.append(varfit)
    varlog_fit_200.append(varlog_fit)
```

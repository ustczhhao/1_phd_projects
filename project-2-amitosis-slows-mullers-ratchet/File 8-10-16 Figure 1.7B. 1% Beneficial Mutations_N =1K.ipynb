```python
generation_list = []

for i in range(1002):
    generation_list.append(i)
    
print(len(generation_list))
```

```python
file_1k = ['Bene01_Amito_Fit_N1K_1KG_NewSex.csv','Bene01_Mito_Fit_N1K_1KG_NewSex.csv', 'Bene01_RM_Fit_N1K_1KG_NewSex.csv',\
           'Bene01_RM_E10_Fit_N1K_1KG_NewSex.csv','Bene01_RM_E100_Fit_N1K_1KG_NewSex.csv']
```

```python
f1k_fit_mean = []
f1k_fit_std = []

for i in file_1k:
    f = pd.read_csv(i,dtype=str)
    
    fit_mean = []
    fit_std = []
       
    fit_mean.extend(list(pd.Series(f.meanFit,dtype=float)))
    fit_std.extend(list(pd.Series(f.PopMeanFit_STD,dtype=float)))
        
    f1k_fit_mean.append(fit_mean)
    f1k_fit_std.append(fit_std)
```

```python
def publication_figure(axis):
    plt.rcParams.update({'font.size': 20})
    ax = axis
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_linewidth(4)
    ax.spines['bottom'].set_linewidth(4)
    ax.spines['left'].set_position(('outward', 10))
    ax.spines['bottom'].set_position(('outward', 10))
    #ax.xaxis.set_ticks([])
    ax.tick_params(axis='x', which='both', bottom='on', top='off',labelbottom='on', size=10, width=5, direction='out', pad=25, colors='black')
    ax.tick_params(axis='y', which='both', left='on', right='off',labelbottom='on', size=10, width=5, direction='out', pad=25, colors='black')
```

```python
fig = plt.figure(figsize=(10, 8), dpi = 300)
ax = fig.add_subplot(111)
publication_figure(ax)
# ax.scatter(0,0)

ax.set_xlim(0,1000)
# ax.set_ylim(-12,0)

t = generation_list

a = f1k_fit_mean[0]
error_a = list(1.96* np.array(f1k_fit_std[0])/((500)**0.5))

b = f1k_fit_mean[1]
error_b = list(1.96* np.array(f1k_fit_std[1])/((500)**0.5))


c = f1k_fit_mean[2]
error_c = list(1.96* np.array(f1k_fit_std[2])/((500)**0.5))


d = f1k_fit_mean[3]
error_d = list(1.96* np.array(f1k_fit_std[3])/((500)**0.5))


e = f1k_fit_mean[4]
error_e = list(1.96* np.array(f1k_fit_std[4])/((500)**0.5))


# ax.set_title('Amitosis vs. mitosis', fontsize =23)
ax.set_xlabel('Time (generations)', fontsize =20)
ax.set_ylabel('Mean Fitness', fontsize =20)


# ax.semilogy(t, a, )

ax.set_yscale('log')

ax.plot(t,a,'blue', label = 'Amitosis',  linewidth=2.0)
ax.fill_between(t, np.array(a)- np.array(error_a), np.array(a)+ np.array(error_a), alpha =0.2, facecolor='blue')

ax.plot(t,b,'deepskyblue', label = 'Mitosis', linewidth=2.0)
ax.fill_between(t, np.array(b)- np.array(error_b), np.array(b)+ np.array(error_b), alpha =0.2, facecolor='deepskyblue')

ax.plot(t,c,'red', label = r"Sex $(\tau=1)$",  linewidth=2.0)
ax.fill_between(t, np.array(c)- np.array(error_c), np.array(c)+ np.array(error_c), alpha =0.2, facecolor='red')

ax.plot(t,d,'green', label = r"Sex $(\tau=10)$",  linewidth=2.0)
ax.fill_between(t, np.array(d)- np.array(error_d), np.array(d)+ np.array(error_d), alpha =0.2, facecolor='green')

ax.plot(t,e,'m', label = r"Sex $(\tau=100)$",  linewidth=2.0)
ax.fill_between(t, np.array(e)- np.array(error_e), np.array(e)+ np.array(error_e), alpha =0.2, facecolor='m')

plt.axhline(y=1, xmin=0, xmax=1000, ls ='dashed', color ='black', linewidth=1.5)

ax.minorticks_off()
# ax.yaxis.set_ticks(np.arange(start, end, 0.1))

plt.legend(loc='upper left', frameon=False, prop={'size':16})

extraticks=[0.8, 1.6]
ax.set_yticks(list(ax.get_yticks()) + extraticks)

plt.ylim(8*10**(-1), 1.6)

# plt.show()
plt.gca().get_xaxis().set_major_formatter(FuncFormatter(lambda x, p: format(int(x), ',')))
```

```python
file_20 = ['T1_Fitness_Amito_20_10K_E1.csv', 'T1_Fitness_Mito_20_10K_E1.csv', 'T1_Fitness_SF_20_10K_E1.csv', \
         'T1_Fitness_RM_20_10K_E1.csv', 'T1_Fitness_SF_E100_20_10K_E1.csv', 'T1_Fitness_RM_E100_20_10K_E1.csv']

mfit_20 = []
log_mfit_20 = []

varfit_20 = []
varlog_fit_20 =[]

for i in file_20:
    f = pd.read_csv(i,dtype=str)
    
    mfit = []
    log_mfit = []
    
    varfit = []
    varlog_fit =[]
    
    mfit.extend(list(pd.Series(f.meanFit,dtype=float)))
    log_mfit.extend(list(pd.Series(f.log_meanFit,dtype=float)))
    
    varfit.extend(list(pd.Series(f.varFit,dtype=float)))
    varlog_fit.extend(list(pd.Series(f.varlog_Fit,dtype=float)))
    
    mfit_20.append(mfit)
    log_mfit_20.append(log_mfit)
    
    varfit_20.append(varfit)
    varlog_fit_20.append(varlog_fit)
```

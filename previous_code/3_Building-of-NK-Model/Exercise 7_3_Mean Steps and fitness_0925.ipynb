```python
aver_step = []
std_step = []
aver_optimal_fitness = []
std_optimal_fitness = []


walk_24_list = [walk_1002, walk_10010, walk_10020]

for i in walk_24_list:
    step_length = []
    optimal_fitness = []
    for j in i:
        for s in j:
            step_length.append(len(s)-1)
            optimal_fitness.append(s[-1][-1])
            
    aver_step_length = sum(step_length)/len(step_length)
    std_step_length = np.std(step_length, ddof = 1)
    aver_fitness = sum(optimal_fitness)/len(optimal_fitness)
    std_fitness = np.std(optimal_fitness, ddof = 1)
    
    aver_step.append(aver_step_length)
    std_step.append(std_step_length)
    aver_optimal_fitness.append(aver_fitness)
    std_optimal_fitness.append(std_fitness)
    
se_step = [i*1.96/10 for i in std_step]
se_optimal_fitness = [i*1.96/10 for i in std_optimal_fitness]

print(aver_step)
print(std_step)
print(se_step)
print(aver_optimal_fitness)
print(std_optimal_fitness)
print(se_optimal_fitness)
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

plt.figure(figsize=(10,8))
m1 = [2,10,20]
n1 = np.array(aver_step)
e1 = np.array(se_step)

plt.errorbar(m1, n1, e1, linestyle='None', marker='^')
plt.xlim(0, 21)
plt.ylim(10, 50)

plt.rcParams.update({'font.size': 16})

plt.title('ADD_STEP LENGTH_N =100')
plt.xlabel('Index')
plt.ylabel('Average steps to reach local optima')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

plt.figure(figsize=(10,8))
m1 = [2,10,20]
n1 = np.array(aver_optimal_fitness)
e1 = np.array(se_optimal_fitness)

plt.errorbar(m1, n1, e1, linestyle='None', marker='^')
plt.xlim(0, 21)
plt.ylim(0.55, 0.75)

plt.rcParams.update({'font.size': 16})

plt.title('ADD_OPTIMAL FITNESS_N =100')
plt.xlabel('Index')
plt.ylabel('Mean Fitness of local optima ')
plt.show()
```

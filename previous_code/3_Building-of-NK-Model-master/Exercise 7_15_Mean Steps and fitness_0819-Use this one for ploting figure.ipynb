```python
# Additive- Random walk
aver_step_add_rw = []
std_step_add_rw = []
aver_optimal_fitness_add_rw = []
std_optimal_fitness_add_rw = []


random_walk_add = [random_walk_200, random_walk_205, random_walk_2010, random_walk_2015, random_walk_2019]

for i in random_walk_add:
    step_length = []
    optimal_fitness = []
    for j in i:
        for s in j:
            step_length.append(len(s)-1)
            optimal_fitness.append(s[-1][-1])
            
    aver_step = sum(step_length)/len(step_length)
    std_step = np.std(step_length, ddof =1)
    aver_fitness =sum(optimal_fitness)/len(optimal_fitness)
    std_fitness =np.std(optimal_fitness, ddof =1)

    aver_step_add_rw.append(aver_step)
    std_step_add_rw.append(std_step)
    aver_optimal_fitness_add_rw.append(aver_fitness)
    std_optimal_fitness_add_rw.append(std_fitness)
    
print('AVER STEP_ADD_RW', aver_step_add_rw)
print('STD STEP', std_step_add_rw)
print('AVER FITNESS', aver_optimal_fitness_add_rw)
print('STD FITNESS', std_optimal_fitness_add_rw)
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

plt.figure(figsize=(10,8))
m1 = np.array(range(1, len(aver_step_add_rw)+1))
n1 = np.array(aver_step_add_rw)
e1 = np.array(std_step_add_rw)

plt.errorbar(m1, n1, e1, linestyle='None', marker='^')
plt.xlim(0, len(aver_step_add_rw)+1)
plt.ylim(0, 15)

plt.rcParams.update({'font.size': 16})

plt.title('Additive_Random Walk_N =20')
plt.xlabel('Index')
plt.ylabel('Average steps to reach local optima')
plt.show()
```

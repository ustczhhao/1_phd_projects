```python
aver_step = []
std_step = []
aver_optimal_fitness = []
std_optimal_fitness = []


total_step_length_40 = []
total_optimal_fitness_40 = []

for i in walk_40:
    for j in i:
        total_step_length_40.append(len(j))
        total_optimal_fitness_40.append(j[-1][-1])
        
aver_step_40 = sum(total_step_length_40)/len(total_step_length_40)
std_step_40 = np.std(total_step_length_40, ddof =1)
aver_fitness_40 =sum(total_optimal_fitness_40)/len(total_optimal_fitness_40)
std_fitness_40 =np.std(total_optimal_fitness_40, ddof =1)

aver_step.append(aver_step_40)
std_step.append(std_step_40)
aver_optimal_fitness.append(aver_fitness_40)
std_optimal_fitness.append(std_fitness_40)

print(aver_step_40, std_step_40, aver_fitness_40, std_fitness_40)
```

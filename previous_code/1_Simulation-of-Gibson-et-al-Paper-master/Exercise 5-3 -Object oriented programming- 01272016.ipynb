```python
class ChemicalSpecies(object):
    
    def __init__(self, name, count):
        self.name=name
        self.count=count
        
        
class Reaction1(object):
    '''A+B-->C, two reactants and one product'''
    
    def __init__(self, reactant1, reactant2, product1, reconstant):
        self.reactant1=reactant1
        self.reactant2=reactant2
        self.product1=product1
        self.reaction_constant=reconstant
        
    def propensity(self):
        self.propensity=self.reaction_constant*self.reactant1.count*self.reactant2.count
        
    
    def execution(self):
        self.reactant1.count=self.reactant1.count-1
        self.reactant2.count=self.reactant2.count-1
        self.product1.count=self.product1.count+1
        
    
    
class Reaction2(object):
    '''A+B-->C+D, two reactants and two products'''
    
    def __init__(self, reactant1, reactant2, product1,product2, reconstant):
        self.reactant1=reactant1
        self.reactant2=reactant2
        self.product1=product1
        self.product2=product2
        self.reaction_constant=reconstant
        
    def propensity(self):
        self.propensity=self.reaction_constant*self.reactant1.count*self.reactant2.count
        
    def execution(self):
        self.reactant1.count=self.reactant1.count-1
        self.reactant2.count=self.reactant2.count-1
        self.product1.count=self.product1.count+1
        self.product2.count=self.product2.count+1
        
        
class Reaction3(object):
    '''A-->B+C, one reactant and two products'''
    
    def __init__(self, reactant1, product1,product2, reconstant):
        self.reactant1=reactant1
        self.product1=product1
        self.product2=product2
        self.reaction_constant=reconstant
        
    def propensity(self):
        self.propensity=self.reaction_constant*self.reactant1.count
        return self.propensity
    
    def execution(self):
        self.reactant1.count=self.reactant1.count-1
        self.product1.count=self.product1.count+1
        self.product2.count=self.product2.count+1
        return self.reactant1.count, self.product1.count, self.product2.count
    
```

```python
a=ChemicalSpecies('A',6)
b=ChemicalSpecies('B',14)
c=ChemicalSpecies('C',8)
d=ChemicalSpecies('D',12)
e=ChemicalSpecies('E',9)
f=ChemicalSpecies('F',3)
g=ChemicalSpecies('G',5)
```

```python
rx1=Reaction1(a,b,c,1)
p1=rx1.propensity()
p2=rx1.execution()
print(p1,p2)
```

```python
rx2=Reaction1(b,c,d,2)
p3=rx2.propensity()
p4=rx2.execution()
print(p3,p4)
```

```python
rx3=Reaction2(d,e,e,f,3)
p5=rx3.propensity()
p6=rx3.execution()
print(p5,p6)
```

```python
rx4=Reaction3(f,d,g,4)
p7=rx4.propensity()
p8=rx4.execution()
print(p7,p8)
```

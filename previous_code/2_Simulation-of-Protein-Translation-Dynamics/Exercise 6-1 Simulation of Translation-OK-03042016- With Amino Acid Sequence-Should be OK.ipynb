```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def propensity(k,*args):
    
    for i in args:
        k=k*i
    return k


def ribosomebinding(a,b,c):
    '''the binding of ribosome into RNA'''
    
    if a>0 and b>0:
        a-=1
        b-=1
        c+=1
    return a,b,c


def releaserbs(a,b,c):
    '''Ribosome read through the second codon, release the RBS of mRNA'''
    
    if a>0:
        a-=1
        b+=1
        c+=1
    return a,b,c


def readthrough(a,b):
    '''ribosome read through mRNA, actually the transition of complex
    or encounter the stop codon, release the free ribsome'''
    
    if a>0:
        a-=1
        b+=1
    return a,b


def translation(sequence):  # how many codons in the sequence
    
    codon_dict={'UUU':'F', 'UUC':'F', 'UUA':'L', 'UUG':'L',     
                 'UCU':'S', 'UCC':'S', 'UCA':'S', 'UCG':'S',     
                 'UAU':'Y', 'UAC':'Y', 'UAA':'STOP', 'UAG':'STOP',     
                 'UGU':'C', 'UGC':'C', 'UGA':'STOP', 'UGG':'W',     
                 'CUU':'L', 'CUC':'L', 'CUA':'L', 'CUG':'L',     
                 'CCU':'P', 'CCC':'P', 'CCA':'P', 'CCG':'P',     
                 'CAU':'H', 'CAC':'H', 'CAA':'Q', 'CAG':'Q',     
                 'CGU':'R', 'CGC':'R', 'CGA':'R', 'CGG':'R',     
                 'AUU':'I', 'AUC':'I', 'AUA':'I', 'AUG':'M',     
                 'ACU':'T', 'ACC':'T', 'ACA':'T', 'ACG':'T',     
                 'AAU':'N', 'AAC':'N', 'AAA':'K', 'AAG':'K',     
                 'AGU':'S', 'AGC':'S', 'AGA':'R', 'AGG':'R',     
                 'GUU':'V', 'GUC':'V', 'GUA':'V', 'GUG':'V',     
                 'GCU':'A', 'GCC':'A', 'GCA':'A', 'GCG':'A',     
                 'GAU':'D', 'GAC':'D', 'GAA':'E', 'GAG':'E',    
                 'GGU':'G', 'GGC':'G', 'GGA':'G', 'GGG':'G'}
    
    aa_list=[]
    
    start_point=sequence.find('AUG')
    trimmed_sequence=sequence[start_point:]
    
    codons_list = [trimmed_sequence[i:i+3] for i in range(0, len(trimmed_sequence), 3)]
    
    stop_list=['UAA', 'UAG','UGA']
    
    n=0
    for i in codons_list:
        if len(i)==1 or len(i)==2:
            pass
        elif i not in stop_list:
            n+=1
            aa_list.append(codon_dict[i])
        else:
            break
            
    return n,aa_list
    

def calculation_constant(sequence):
    
    codonconstant_dict={'UUU':1.01, 'UUC':1.02, 'UUA':1.03, 'UUG':1.04,     
                        'UCU':1.05, 'UCC':1.06, 'UCA':1.07, 'UCG':1.08,     
                        'UAU':1.09, 'UAC':1.10, 'UAA':'STOP', 'UAG':'STOP',     
                        'UGU':1.11, 'UGC':1.12, 'UGA':'STOP', 'UGG':1.14,     
                        'CUU':1.15, 'CUC':1.16, 'CUA':1.17, 'CUG':1.18,     
                        'CCU':1.19, 'CCC':1.20, 'CCA':1.21, 'CCG':1.22,     
                        'CAU':1.23, 'CAC':1.24, 'CAA':1.25, 'CAG':1.26,     
                        'CGU':1.27, 'CGC':1.28, 'CGA':1.29, 'CGG':1.30,     
                        'AUU':1.31, 'AUC':1.32, 'AUA':1.33, 'AUG':1.34,     
                        'ACU':1.35, 'ACC':1.36, 'ACA':1.37, 'ACG':1.38,     
                        'AAU':1.39, 'AAC':1.40, 'AAA':1.41, 'AAG':1.42,     
                        'AGU':1.43, 'AGC':1.44, 'AGA':1.45, 'AGG':1.46,     
                        'GUU':1.47, 'GUC':1.48, 'GUA':1.49, 'GUG':1.50,     
                        'GCU':1.51, 'GCC':1.52, 'GCA':1.53, 'GCG':1.54,     
                        'GAU':1.55, 'GAC':1.56, 'GAA':1.57, 'GAG':1.58,    
                        'GGU':1.59, 'GGC':1.60, 'GGA':1.61, 'GGG':1.62}
    
    
    reconstant_list=[]
    
    start_point=sequence.find('AUG')
    trimmed_sequence=sequence[start_point:]
    
    codons_list = [trimmed_sequence[i:i+3] for i in range(0, len(trimmed_sequence), 3)]
    
    stop_list=['UAA', 'UAG','UGA']
    
    for i in codons_list:
        if len(i)==1 or len(i)==2:
            pass
        elif i not in stop_list:
            reconstant_list.append(codonconstant_dict[i])
        else:
            break
    
    return reconstant_list
            

def directmethod(codon_number,aa_list, complex_list,reconstant_list, total_rib, rna, t1, steps): 

    n=codon_number
    
    rib=total_rib
    rna_rbs=rna
    
    prop_list=[]
    prop1=propensity(reconstant_list[0],rib,rna_rbs)
    
    prop_list.append(prop1)
         
    for i in range(1,n):
        prop_list.append(propensity(reconstant_list[i],complex_list[i-1]))
    
    if steps:
        new_prop = []
        for i, p in enumerate(prop_list):
            if i not in steps:
                new_prop.append(p)
            else:
                new_prop.append(0)
    else:
        new_prop = prop_list
    
#     print('STEPS IN　DIR',steps)
#     print('PROP',prop_list)
#     print('NEW PROP', new_prop)
    
    sumprop=sum(new_prop)
    prob_list=[]

    if sumprop > 0:
        for i in new_prop: 
            prob_list.append(i/sumprop)
    
    print('PROP',prop_list)
    
    print('prop1',prop1, 'RIB',rib, 'RNA_RBS',rna_rbs)
    print('NEW PROP',new_prop)
    print('Prob_list',prob_list)
    
    if sum(prob_list)!=0:
        u=random.choice(np.arange(0,n),p=prob_list)  
        tau=random.exponential(1/sumprop)
        
        if u==0:
#             print('Reaction start', t1)
            rib,rna_rbs,complex_list[0]=ribosomebinding(rib,rna_rbs,complex_list[0])  
#             print(rib,rna_rbs,complex_list[0])
        elif u==1:
#             print('First position',t1)
            complex_list[0],complex_list[1],rna_rbs=releaserbs(complex_list[0],complex_list[1],rna_rbs) # release the RBS
        elif u==n-1:
#             print('Stop',t1)
            complex_list[u-1],rib=readthrough(complex_list[u-1],rib) # encouter the STOP codon
        else:
#             print('Elongation', u,t1)
            complex_list[u-1],complex_list[u]=readthrough(complex_list[u-1],complex_list[u])
        
    newcomplex_list=[]  
    for i in range(len(complex_list)):
        newcomplex_list.append(complex_list[i])
#         print('NEWCOMP', newcomplex_list)
        
    rib1=rib
    rna_rbs1=rna_rbs
    
    aa=aa_list[u]
                
    return newcomplex_list, rib1, rna_rbs1, tau, aa, u
        
        

def simulation(sequence,total_rib,t1,t2):
    
    t0=t1
    t1_list=[]
    
    s=translation(sequence)
    n=s[0]
    aa_list=s[1]
    
    reconstant_list=calculation_constant(sequence)
    
    complex_list=[0]*(n-1)
     
    rib=total_rib
    
    rna_rbs=1
    
    result={}
    aa_result={}
    
    ids=0
    u_list=[]
    
    steps=[]

    while t1<t2:
#         print(t1)
        y=directmethod(n,aa_list, complex_list,reconstant_list, rib,rna_rbs,t1, steps)
        u=y[-1]
        aa=y[-2]
        
        print('U is',u)
        
        if u_list==[]:
            complex_list=y[0]
            rib=y[1]
            rna_rbs=y[2]

            t1+=y[3]
            t1_list.append(t1)
            
            u_list.append(u)
            ids+=1
            result[ids]=[(t1,u)]
            aa_result[ids]=[aa]
            
            steps=[u]
            
            print('Initial Steps', steps)
            
        else:
#             steps = [v[-1][-1] for v in result.values() if v[-1][-1]!=n-1]
#             steps = [v - 1 for v in steps if v!=0] + steps

            complex_list=y[0]
            rib=y[1]
            rna_rbs=y[2]

            u_list.append(u)
                
            if u==0:
                ids+=1
                t1 += y[3]
                result[ids]=[(t1,u)]
                aa_result[ids]=[aa]
            else:
                t1 += y[3]
                for k, v in result.items():
                    if v[-1][-1]==u-1:
                        result[k].append((t1, u))
                        aa_result[k].append(aa)
                        break

            steps = [v[-1][-1] for v in result.values() if v[-1][-1]!=n-1]
            steps = [v - 1 for v in steps if v!=0] + steps   
            print('STEPS is', steps)
    
    for k,v in aa_result.items():
#         print('Reaction {}:'.format(k),':' ,''.join(v))
        print('Reaction {}: {}'.format(k,''.join(v)))        
    
    return result,aa_result
       
```

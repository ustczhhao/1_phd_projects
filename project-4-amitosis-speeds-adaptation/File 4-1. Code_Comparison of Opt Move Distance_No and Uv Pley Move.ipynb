```python
'''

As in our deterministic moving of optimum, the no pleiotropy and universal pleiotropy movement of optimum with the same moving
STD caused different results (here no pleiotrpy moving means that each movement of optimum only affects its corresponding 
dimension, universal pleiotropy moving means that each movement of optimum will affect all the dimension).

We thought with STD = 0.01, the distance of each movement under unviersal pleiotropy may move the optimum in a much short distance
than that in the no pleiotropy. Here is the code to calculate the movement distance of optimum under no and universal pleiotropy.


'''
```

### No Pleiotropy Movement

```python
from __future__ import division
import numpy as np
from scipy import stats
import scipy.spatial as spa
import numpy.random as rnd
import copy
import time
import pandas as pd
import math
import pickle
```

```python
class Optimum_move1(object):
    '''Define a single population in FGM. Assuming the fixed optimum is at the orgin of the Coordinate System'''

    def __init__(self,nReps, dim, optimum_move_sigma):
        
        self.nReps = nReps
        self.dim = dim  # The dimension of the landscape

        self.soma_optimum = np.zeros((nReps, dim))
        self.opt_move_sigma = optimum_move_sigma

        self.current_step =0  # Will be used to indicate the number of generations that have been simulated.
        self.move_distance = 0


    def move_optimum(self):

        soma_opt_mut = np.random.normal(0, self.opt_move_sigma, size = self.nReps*self.dim).reshape(self.nReps, self.dim)

        self.move_distance = np.sqrt(np.sum((soma_opt_mut)**2, axis =1)) 
        

        self.soma_optimum = self.soma_optimum + soma_opt_mut

        self.current_step +=1




    def calculate_distance(self):

        D = self.move_distance

        D_mean = np.nanmean(D)
        D_std = np.nanstd(D)
        D_var = np.var(D)

        return [D_mean, D_std, D_var]




    def simulate(self,stepcount):

        distance = [self.calculate_distance()]

        start = time.time()

        while self.current_step <= stepcount:
            self.move_optimum()
            distance.append(self.calculate_distance())


        colnames = ['OptMoveDistance_Mean','OptMoveDistance_Std','OptMoveDistance_Var']
        distance = pd.DataFrame(np.array(distance),columns=colnames)


        end = time.time()
        print 'TOTAL TIME: ',end-start

        return distance


    def simulateNsave(self,outfile,stepcount,):
        """same as simulate method except results are written to the file specified by outfile"""
        results = self.simulate(stepcount)
        results.to_csv(outfile,index=False)  
        
        return    
```

```python
a = Optimum_move1(nReps =100, dim =10, optimum_move_sigma =0.01)
a.simulateNsave('NoPley_Move.csv', 1000)
```

### Universal Pleiotropy Movement

```python
from __future__ import division
import numpy as np
from scipy import stats
import scipy.spatial as spa
import numpy.random as rnd
import copy
import time
import pandas as pd
import math
import pickle
```

```python
class Optimum_move2(object):
    '''Define a single population in FGM. Assuming the fixed optimum is at the orgin of the Coordinate System'''

    def __init__(self,nReps, dim, optimum_move_sigma):
        
        self.nReps = nReps
        self.dim = dim  # The dimension of the landscape

        self.soma_optimum = np.zeros((nReps, dim))
        self.opt_move_sigma = optimum_move_sigma/np.sqrt(dim)
        self.opt_move_matrix = np.identity(self.dim)*(self.opt_move_sigma**2)

        self.current_step =0  # Will be used to indicate the number of generations that have been simulated.
        self.move_distance = 0


    def move_optimum(self):

        soma_opt_mut = np.random.multivariate_normal(mean=np.zeros(self.dim),cov=self.opt_move_matrix,size= self.nReps)

        self.move_distance = np.sqrt(np.sum((soma_opt_mut)**2, axis =1)) 
        
        self.soma_optimum = self.soma_optimum + soma_opt_mut

        self.current_step +=1




    def calculate_distance(self):

        D = self.move_distance

        D_mean = np.nanmean(D)
        D_std = np.nanstd(D)
        D_var = np.var(D)

        return [D_mean, D_std, D_var]




    def simulate(self,stepcount):

        distance = [self.calculate_distance()]

        start = time.time()

        while self.current_step <= stepcount:
            self.move_optimum()
            distance.append(self.calculate_distance())


        colnames = ['OptMoveDistance_Mean','OptMoveDistance_Std','OptMoveDistance_Var']
        distance = pd.DataFrame(np.array(distance),columns=colnames)


        end = time.time()
        print 'TOTAL TIME: ',end-start

        return distance


    def simulateNsave(self,outfile,stepcount,):
        """same as simulate method except results are written to the file specified by outfile"""
        results = self.simulate(stepcount)
        results.to_csv(outfile,index=False)  
        
        return    
```

```python
a = Optimum_move2(nReps =100, dim =10, optimum_move_sigma =0.01)
a.simulateNsave('UvPley_Move.csv', 1000)
```

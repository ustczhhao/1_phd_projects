```python
from __future__ import division
import matplotlib as lib
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
import pylab
import pandas as pd
import os
import pickle
%matplotlib inline
```

```python
generation_list = []

for i in range(2002):
    generation_list.append(i)
    
print(len(generation_list))
```

```python
def get_data1(file_list):
    
    total_fit_mean = []
    total_fit_std = []
    
    total_var_mean = []
    total_var_std = []
    
    for i in file_list:
        f = pd.read_csv(i,dtype=str)
    
        fit_mean = []
        fit_std = []
    
        var_mean = []
        var_std = []
       
        fit_mean.extend(list(pd.Series(f.mFit_Mean,dtype=float)))
        fit_std.extend(list(pd.Series(f.mFit_Std,dtype=float)))
    
        var_mean.extend(list(pd.Series(f.varFit_Mean,dtype=float)))
        var_std.extend(list(pd.Series(f.varFit_Std,dtype=float)))
    
        total_fit_mean.append(fit_mean)
        total_fit_std.append(fit_std)
    
        total_var_mean.append(var_mean)
        total_var_std.append(var_std)  
        
    return total_fit_mean, total_fit_std, total_var_mean, total_var_std
```

```python
w_mito = ['Cor_Rv_MoveOpt_SM006_NoPley_GMU001_Dim10_P45_Mito_Fit_N1K_Fit05_SM005_2KG.csv']
```

```python
w_s = get_data1(w_mito)

w_mito_fit_mean =w_s[0]
w_mito_fit_std = w_s[1]
w_mito_var_mean =w_s[2]
w_mito_var_std = w_s[3]
```

```python
def publication_figure(axis):
    plt.rcParams.update({'font.size': 36})
    ax = axis
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_linewidth(6)
    ax.spines['bottom'].set_linewidth(6)
    ax.spines['left'].set_position(('outward', 10))
    ax.spines['bottom'].set_position(('outward', 10))
    #ax.xaxis.set_ticks([])
    ax.tick_params(axis='x', which='both', bottom='on', top='off',labelbottom='on', size=20, width=6, direction='out', pad=50, colors='black')
    ax.tick_params(axis='y', which='both', left='on', right='off',labelbottom='on', size=20, width=6, direction='out', pad=50, colors='black')
```

```python
fig = plt.figure(figsize=(16, 12),dpi =300)
ax = fig.add_subplot(111)
publication_figure(ax)
# ax.scatter(0,0)

ax.set_xlim(0,2000)

t = generation_list


g1 = w_amito_fit_mean[0]
error_g1 = list(1.98* np.array(w_amito_fit_std[0])/((100)**0.5))

g2 = w_mito_fit_mean[0]
error_g2 = list(1.98* np.array(w_mito_fit_std[0])/((100)**0.5))


# ax.set_title('FGM_No Pley_Moving Opt_MO Deterministic and STD =0.01_Corrected', fontsize =35)
ax.set_xlabel('Time (generations)', fontsize =36)
ax.set_ylabel('Population Mean Fitness', fontsize =36)



ax.plot(t,g1,'b', label = 'Amitosis',  linewidth=2.0, alpha =1)
ax.fill_between(t, np.array(g1)- np.array(error_g1), np.array(g1)+ np.array(error_g1), alpha =0.3, facecolor='b')

ax.plot(t,g2,'deepskyblue', label = 'Mitosis',   linewidth=2.0, alpha =1)
ax.fill_between(t, np.array(g2)- np.array(error_g2), np.array(g2)+ np.array(error_g2), alpha =0.3, facecolor='deepskyblue')




plt.axhline(y=0.5, linewidth=2.0, color ='k',ls ='dashed')

ax.minorticks_off()

ax.set_ylim(0, 1)

# plt.legend(loc='upper left', frameon=False, prop={'size':28})
```

```python
'''
This is corrected version of Probabilitic Movement of optimum. 

Here we simulae the evolution under moving optimum. In this simulation, the optimum will move according to a similar strategy as
generating mutations, i.e., the optimum will move along each dimension according to normal distribution, with a given total 
moving rate and a pre-set STD. 

Here we assumed no pleiotropy. The moving rate of optimum is 1 each generation, and the STD of moving is 0.01.

Here the plotting used the corrected data collected with adding random.seed().


'''
```

```python
from __future__ import division
import matplotlib as lib
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
import pylab
import pandas as pd
import os
import pickle
%matplotlib inline
```

```python
generation_list = []

for i in range(2002):
    generation_list.append(i)
    
print(len(generation_list))
```

```python
def get_data1(file_list):
    
    total_fit_mean = []
    total_fit_std = []
    
    total_var_mean = []
    total_var_std = []
    
    for i in file_list:
        f = pd.read_csv(i,dtype=str)
    
        fit_mean = []
        fit_std = []
    
        var_mean = []
        var_std = []
       
        fit_mean.extend(list(pd.Series(f.mFit_Mean,dtype=float)))
        fit_std.extend(list(pd.Series(f.mFit_Std,dtype=float)))
    
        var_mean.extend(list(pd.Series(f.varFit_Mean,dtype=float)))
        var_std.extend(list(pd.Series(f.varFit_Std,dtype=float)))
    
        total_fit_mean.append(fit_mean)
        total_fit_std.append(fit_std)
    
        total_var_mean.append(var_mean)
        total_var_std.append(var_std)  
        
    return total_fit_mean, total_fit_std, total_var_mean, total_var_std
```
